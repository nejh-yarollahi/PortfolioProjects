#### Combining our code to crawl through all pages

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

main_r = requests.get("https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s=0.html")
main_c = main_r.content
main_soup = BeautifulSoup(main_c, "html.parser")
page_nr = main_soup.find_all("a", {"class": "Page"})[-1].text

l=[]
base_url = "https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s="

for page in range(0, int(page_nr) * 10, 10):
    print(base_url + str(page) + ".html")
    r = requests.get(base_url + str(page) + ".html")
    c = r.content
    soup = BeautifulSoup(c, 'html.parser')
    all_properties = soup.find_all("div", {"class": "propertyRow"})

    for prop in all_properties:
        d = {}
        d["Address"] = prop.find_all("span", {"class": "propAddressCollapse"})[0].text
        d["Locality"] = prop.find_all("span", {"class": "propAddressCollapse"})[1].text
        d["Price"] = prop.find("h4", {"class": "propPrice"}).text.replace("\n", "").replace(" ", "")
        try:
            d["Beds"] = prop.find("span", {"class": "infoBed"}).find("b").text
        except:
            d["Beds"] = None
        try:
            d["Area"] = prop.find("span", {"class": "infoSqFt"}).find("b").text
        except:
            d["Area"] = None
        try:
            d["Full Baths"] = prop.find("span", {"class": "infoValueFullBath"}).find("b").text
        except:
            d["Full Baths"] = None
        try:
            d["Half Baths"] = prop.find("span", {"class": "infoValueHalfBath"}).find("b").text
        except:
            d["Half Baths"] = None
        for column_group in prop.find_all('div', {"class": "columnGroup"}):
            # print(column_group)
            for feature_group, feature_name in zip(column_group.find_all("span", {"class": "featureGroup"}),
                                                column_group.find_all("span", {"class": "featureName"})):
                # print(feature_group.text, feature_name.text)
                if "Lot Size" in feature_group.text:
                    d["Lot Size"] = feature_name.text
        l.append(d)
```

#### Saving our data in a dataframe

```python
df = pd.DataFrame(l)
l
```

#### Saving our data on a csv file

```python
df.to_csv("century21_output.csv")
```
